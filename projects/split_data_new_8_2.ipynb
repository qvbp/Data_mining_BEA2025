```python
# 打开原始数据集
with open('/mnt/cfs/huangzhiwei/BAE2025/data/mrbench_v3_devset.json', 'r', encoding='utf-8') as f:
    datas = json.load(f)
    
new_data = []
for data in datas:
    experts = data["tutor_responses"]
    # print(type(experts))
    # print("专家数量：", len(experts))
    
    # 修改代码部分，针对标注处理的逻辑
    for expert_name, expert_data in experts.items():
        # 获取专家回答和标注
        response = expert_data["response"]
        labels = expert_data["annotation"]
        
        # 为每种标注类型创建一个字典
        data_dict = {
            "conversation_history": data["conversation_history"],
            "response": response,
            "Mistake_Identification": labels.get("Mistake_Identification", ""),  # 使用get方法安全获取值
            "Mistake_Location": labels.get("Mistake_Location", ""),
            "Providing_Guidance": labels.get("Providing_Guidance", ""),
            "Actionability": labels.get("Actionability", "")
        }
        
        new_data.append(data_dict)
            
# 将数据集保存为json文件
with open('/mnt/cfs/huangzhiwei/BAE2025/data_new/all.json', 'w', encoding='utf-8') as f:
    json.dump(new_data, f, ensure_ascii=False, indent=4)
    
# 按照8:2的比例划分训练集和验证集，对每一个json文件
import random
def split_data(data, train_ratio=0.8):
    random.shuffle(data)
    split_index = int(len(data) * train_ratio)
    return data[:split_index], data[split_index:]

# 读取数据
with open('/mnt/cfs/huangzhiwei/BAE2025/data_new/all.json', 'r', encoding='utf-8') as f:
    data = json.load(f)
    
train_data, val_data = split_data(data)
# 保存训练集和验证集
with open('/mnt/cfs/huangzhiwei/BAE2025/data_new/train.json', 'w', encoding='utf-8') as f:
    json.dump(train_data, f, ensure_ascii=False, indent=4)
    
with open('/mnt/cfs/huangzhiwei/BAE2025/data_new/val.json', 'w', encoding='utf-8') as f:
    json.dump(val_data, f, ensure_ascii=False, indent=4)

print("数据集划分完成！")
                        
            
        
        
        

        
```
