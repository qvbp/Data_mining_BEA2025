```python
import pandas as pd
import numpy as np
from itertools import product
import matplotlib.pyplot as plt
import seaborn as sns
import json
import re

# 设置中文显示
plt.rcParams['font.sans-serif'] = ['SimHei']  # 用来正常显示中文标签
plt.rcParams['axes.unicode_minus'] = False  # 用来正常显示负号

# 定义类别和指标
possible_values = ["Yes", "No", "To some extent"]
indicators = ["Mistake_Identification", "Mistake_Location", "Providing_Guidance", "Actionability"]

# 从JSON文件加载数据的函数
def load_data_from_json(file_path):
    """
    从JSON文件加载数据并提取所有response的annotation
    """
    with open(file_path, 'r', encoding='utf-8') as f:
        # 读取文件内容
        content = f.read()
        
        # 修复JSON格式问题（如果有）
        if content.endswith(','):
            content = content + "{}]}"
        elif not content.endswith(']}'):
            content = content + "]}"
        
        try:
            # 尝试解析完整JSON
            data = json.loads(content)
        except json.JSONDecodeError:
            # 如果完整解析失败，使用正则表达式提取注释部分
            return extract_annotations_with_regex(content)
            
        # 提取所有response的annotation
        annotations = []
        
        # 遍历conversation_id下的数据
        for item in data:
            if 'tutor_responses' in item:
                # 遍历每个模型的response
                for model_name, response_data in item['tutor_responses'].items():
                    if 'annotation' in response_data and response_data['annotation']:
                        annotations.append(response_data['annotation'])
        
        return annotations

# 使用正则表达式从文件内容提取注释
def extract_annotations_with_regex(content):
    """
    当JSON解析失败时，使用正则表达式提取annotation数据
    """
    annotations = []
    pattern = r'"annotation":\s*{\s*"Mistake_Identification":\s*"([^"]+)",\s*"Mistake_Location":\s*"([^"]+)",\s*"Providing_Guidance":\s*"([^"]+)",\s*"Actionability":\s*"([^"]+)"'
    
    matches = re.findall(pattern, content)
    for match in matches:
        mi, ml, pg, act = match
        annotations.append({
            "Mistake_Identification": mi,
            "Mistake_Location": ml,
            "Providing_Guidance": pg,
            "Actionability": act
        })
    
    return annotations

# 加载数据
# 将'your_json_file.json'替换为你的实际文件名
file_path = '/mnt/cfs/huangzhiwei/BAE2025/data/mrbench_v3_devset.json'  # 你的JSON文件路径
data = load_data_from_json(file_path)

# 创建数据框
df = pd.DataFrame(data)
print("原始数据:")
print(df.head())
print(f"总样本数: {len(df)}")

# ======== 分析所有81种组合 ========

# 1. 生成所有81种可能组合
all_combinations = []
for values in product(possible_values, repeat=len(indicators)):
    combination = {}
    for i, indicator in enumerate(indicators):
        combination[indicator] = values[i]
    all_combinations.append(combination)

all_combinations_df = pd.DataFrame(all_combinations)
print(f"\n可能的组合总数: {len(all_combinations_df)}")  # 应该是81

# 2. 计算实际数据中每种组合的频率
combination_counts = df.groupby(indicators).size().reset_index(name='count')
print(f"\n实际数据中发现的唯一组合数: {len(combination_counts)}")

# 3. 创建包含所有81种组合及其计数的完整数据框
full_counts = pd.merge(
    all_combinations_df, 
    combination_counts, 
    on=indicators, 
    how='left'
).fillna(0)

# 按计数排序以便查看
sorted_counts = full_counts.sort_values('count', ascending=False)
print("\n按频率排序的组合(前10个):")
print(sorted_counts.head(10))

# 4. 获取缺失组合的统计信息
missing_combinations = full_counts[full_counts['count'] == 0]
print(f"\n缺失的组合: {len(missing_combinations)} 个（共81个）")

# ======== 可视化和深入分析 ========

# 5. 每个指标的分布
print("\n各指标的分布:")
for indicator in indicators:
    print(f"\n{indicator}:")
    print(df[indicator].value_counts())
    
    # 为该指标创建可视化
    plt.figure(figsize=(8, 5))
    ax = sns.countplot(data=df, x=indicator)
    plt.title(f"{indicator}的分布")
    plt.tight_layout()
    
    # 添加计数标签
    for p in ax.patches:
        ax.annotate(f'{int(p.get_height())}', 
                   (p.get_x() + p.get_width() / 2., p.get_height()),
                   ha = 'center', va = 'bottom')
    
    plt.savefig(f'{indicator.lower()}_distribution.png')

# 6. 相关性分析（将类别值转换为数值）
print("\n相关性分析:")
# 将分类值转换为数值以进行相关性分析
value_map = {"Yes": 2, "To some extent": 1, "No": 0}
df_numeric = df.replace(value_map)
correlation = df_numeric[indicators].corr()
print(correlation)

# 可视化相关性
plt.figure(figsize=(10, 8))
sns.heatmap(correlation, annot=True, cmap='coolwarm', vmin=-1, vmax=1)
plt.title("指标间的相关性")
plt.tight_layout()
plt.savefig('correlation_heatmap.png')

# 7. 当一个指标具有特定值时的模式
print("\n=== 条件分布 ===")
for value in possible_values:
    for base_indicator in indicators:
        filtered_df = df[df[base_indicator] == value]
        if len(filtered_df) > 0:
            print(f"\n当 {base_indicator} 为 '{value}' 时 ({len(filtered_df)} 个样本):")
            for other_indicator in indicators:
                if other_indicator != base_indicator:
                    print(f"  {other_indicator} 的分布:")
                    print(f"  {filtered_df[other_indicator].value_counts().to_dict()}")

# 8. 创建热图显示出现的组合分布
# 重点关注MI和ML的组合
mi_ml_counts = df.groupby(['Mistake_Identification', 'Mistake_Location']).size().unstack(fill_value=0)
plt.figure(figsize=(10, 8))
sns.heatmap(mi_ml_counts, annot=True, cmap='YlGnBu', fmt='g')
plt.title('MI和ML组合分布')
plt.tight_layout()
plt.savefig('mi_ml_heatmap.png')

# ML和PG的组合
ml_pg_counts = df.groupby(['Mistake_Location', 'Providing_Guidance']).size().unstack(fill_value=0)
plt.figure(figsize=(10, 8))
sns.heatmap(ml_pg_counts, annot=True, cmap='YlGnBu', fmt='g')
plt.title('ML和PG组合分布')
plt.tight_layout()
plt.savefig('ml_pg_heatmap.png')

# PG和ACT的组合
pg_act_counts = df.groupby(['Providing_Guidance', 'Actionability']).size().unstack(fill_value=0)
plt.figure(figsize=(10, 8))
sns.heatmap(pg_act_counts, annot=True, cmap='YlGnBu', fmt='g')
plt.title('PG和ACT组合分布')
plt.tight_layout()
plt.savefig('pg_act_heatmap.png')

# 9. 为所有81种组合创建统计表格
combo_stats = pd.DataFrame(columns=['组合编号', '组合详情', '计数', '百分比'])
total_samples = len(df)

# 遍历所有可能组合
for i, combo in enumerate(all_combinations):
    # 构建组合详情字符串
    combo_str = ", ".join([f"{k}: {v}" for k, v in combo.items()])
    
    # 找出当前组合在数据中的计数
    count_row = full_counts[
        (full_counts['Mistake_Identification'] == combo['Mistake_Identification']) & 
        (full_counts['Mistake_Location'] == combo['Mistake_Location']) & 
        (full_counts['Providing_Guidance'] == combo['Providing_Guidance']) & 
        (full_counts['Actionability'] == combo['Actionability'])
    ]
    
    count = count_row['count'].values[0] if len(count_row) > 0 else 0
    percentage = (count / total_samples * 100) if total_samples > 0 else 0
    
    # 添加到统计表格
    combo_stats = pd.concat([combo_stats, pd.DataFrame([{
        '组合编号': i + 1,
        '组合详情': combo_str,
        '计数': count,
        '百分比': percentage
    }])], ignore_index=True)

# 按计数排序
combo_stats = combo_stats.sort_values('计数', ascending=False)

# 将结果保存到CSV文件
combo_stats.to_csv('81种组合统计.csv', index=False, encoding='utf-8-sig')
print(f"\n已将所有81种组合的统计结果保存到 '81种组合统计.csv'")

# 10. 计算每个指标的整体分布
indicator_distributions = {}
for indicator in indicators:
    indicator_distributions[indicator] = df[indicator].value_counts(normalize=True) * 100

# 创建一个汇总表格
summary_df = pd.DataFrame(indicator_distributions)
print("\n各指标的整体分布百分比:")
print(summary_df)

# 将指标分布汇总表保存为CSV
summary_df.to_csv('指标分布汇总.csv', encoding='utf-8-sig')
print("已将指标分布汇总保存到 '指标分布汇总.csv'")

print("\n分析完成！请查看生成的可视化和CSV文件。")
```
