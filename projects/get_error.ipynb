```python
import json
import re

# 读取源文件
with open("/mnt/cfs/huangzhiwei/BAE2025/projects/checkpoints_error/bert-base-uncased_fp_b16_e50_len512_lr1e-05/error_analysis/error_examples_epoch_3.json", "r", encoding="utf-8") as f:
    datas = json.load(f)

# 创建新列表
data_list = []

# 处理每条数据
for data in datas:
    data_dict = {
        "conversation_history": "",
        "tutor_responses": {
            "Sonnet": {
                "response": "",
                "annotation": {
                    "Providing_Guidance": ""
                }
            }
        }
    }
    
    # 关键修复：在处理数据前先对 "\n" 周围的空格进行处理
    # 根据原始样本，应该把 "\n" 保留为字面字符，而不是处理成实际换行符
    
    if "first_sentence" in data:
        # 使用正则表达式将 "\n" 周围的非破空格替换回普通空格
        text = data["first_sentence"]
        text = text.replace("\u00a0\n\u00a0", " \n ")  # 替换 "\n" 周围的非破空格
        text = text.replace("\u00a0", " ")  # 替换所有剩余的非破空格
        data_dict["conversation_history"] = text
    
    if "second_sentence" in data:
        text = data["second_sentence"]
        text = text.replace("\u00a0\n\u00a0", " \n ")
        text = text.replace("\u00a0", " ")
        data_dict["tutor_responses"]["Sonnet"]["response"] = text
    
    if "true_label" in data:
        text = data["true_label"]
        text = text.replace("\u00a0\n\u00a0", " \n ")
        text = text.replace("\u00a0", " ")
        data_dict["tutor_responses"]["Sonnet"]["annotation"]["Providing_Guidance"] = text
    
    data_list.append(data_dict)

# 写入新文件，确保正确处理字符
with open("/mnt/cfs/huangzhiwei/BAE2025/projects/checkpoints_error/bert-base-uncased_fp_b16_e50_len512_lr1e-05/error_analysis/error_1.json", "w", encoding="utf-8") as f:
    json.dump(data_list, f, indent=4, ensure_ascii=False)

print("处理完成，已修复 \\n 周围的非破空格问题！")
```

```python
import json
import re

# 读取源文件
with open("/mnt/cfs/huangzhiwei/BAE2025/projects/checkpoints_error/bert-base-uncased_fp_b16_e50_len512_lr1e-05/error_analysis/new_data copy.json", "r", encoding="utf-8") as f:
    datas = json.load(f)

# 创建新列表
data_list = []

# 处理每条数据
for data in datas:
    data_dict1 = {
        "conversation_history": "",
        "tutor_responses": {
            "Sonnet": {
                "response": "",
                "annotation": {
                    "Providing_Guidance": ""
                }
            }
        }
    }
    
    data_dict2 = {
        "conversation_history": "",
        "tutor_responses": {
            "Sonnet": {
                "response": "",
                "annotation": {
                    "Providing_Guidance": ""
                }
            }
        }
    }
    
    # 关键修复：在处理数据前先对 "\n" 周围的空格进行处理
    # 根据原始样本，应该把 "\n" 保留为字面字符，而不是处理成实际换行符
    
    if "conversation_history" in data:
        # 使用正则表达式将 "\n" 周围的非破空格替换回普通空格
        text = data["conversation_history"]
        text = text.replace("\u00a0\n\u00a0", " \n ")  # 替换 "\n" 周围的非破空格
        text = text.replace("\u00a0", " ")  # 替换所有剩余的非破空格
        data_dict1["conversation_history"] = text
        data_dict2["conversation_history"] = text
    
    if "response1" in data["tutor_responses"]["Sonnet"]:
        text = data["tutor_responses"]["Sonnet"]["response1"]
        text = text.replace("\u00a0\n\u00a0", " \n ")
        text = text.replace("\u00a0", " ")
        data_dict1["tutor_responses"]["Sonnet"]["response"] = text
    
    if "response2" in data["tutor_responses"]["Sonnet"]:
        text = data["tutor_responses"]["Sonnet"]["response2"]
        text = text.replace("\u00a0\n\u00a0", " \n ")
        text = text.replace("\u00a0", " ")
        data_dict2["tutor_responses"]["Sonnet"]["response"] = text
    
    if "annotation" in data["tutor_responses"]["Sonnet"]:
        text = data["tutor_responses"]["Sonnet"]["annotation"]["Providing_Guidance"]
        text = text.replace("\u00a0\n\u00a0", " \n ")
        text = text.replace("\u00a0", " ")
        data_dict1["tutor_responses"]["Sonnet"]["annotation"]["Providing_Guidance"] = text
        data_dict2["tutor_responses"]["Sonnet"]["annotation"]["Providing_Guidance"] = text
    
    data_list.append(data_dict1)
    data_list.append(data_dict2)

# 写入新文件，确保正确处理字符
with open("/mnt/cfs/huangzhiwei/BAE2025/data/error_1_8+8.json", "w", encoding="utf-8") as f:
    json.dump(data_list, f, indent=4, ensure_ascii=False)

print("处理完成，已修复 \\n 周围的非破空格问题！")
```
